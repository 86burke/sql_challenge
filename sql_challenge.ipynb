```python
import pandas as pd
import os
import psycopg2
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
DB_HOST = os.getenv('DB_HOST')
DB_NAME = os.getenv('DB_NAME')
DB_USER = os.getenv('DB_USER')
DB_PASS = os.getenv('DB_PASS')
```

```python
dept_df = pd.read_csv('data/departments.csv')
dept_emp_df = pd.read_csv('data/dept_emp.csv')
dept_mgr_df = pd.read_csv('data/dept_manager.csv')
emps_df = pd.read_csv('data/employees.csv')
salary_df = pd.read_csv('data/salaries.csv')
titles_df = pd.read_csv('data/titles.csv')
```

```python
engine = create_engine(f'postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:5432/HR_db')
```

```python
dept_df.to_sql('departments',engine,index=False,if_exists='append')
print('departments')
```

```python
dept_emp_df.to_sql('dept_emp',engine,index=False,if_exists='append')
print('dept_emp')
```

```python
dept_mgr_df.to_sql('dept_manager',engine,index=False,if_exists='append')
print('dept_manager')
```

```python
emps_df.to_sql('employees',engine,index=False,if_exists='append')
print('employees')
```

```python
salary_df.to_sql('salaries',engine,index=False,if_exists='append')
print('salaries')
```

```python
titles_df.to_sql('titles',engine,index=False,if_exists='append')
print('titles')
```

```python
dept_df.to_sql('dept_emp',engine,index=False,if_exists='append')
print('dept_emp')
```
